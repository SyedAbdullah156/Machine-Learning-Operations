```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
# Synthetic dataset: 1000 samples, each input has 20 features
# Each sample can belong to 5 different classes at once
num_samples = 1000
num_features = 20
num_classes = 5

# Random input data
X = np.random.rand(num_samples, num_features)

# Random multi-label outputs (0/1 per class)
y = np.random.randint(0, 2, size=(num_samples, num_classes))
```

```python
# Define the model
model = models.Sequential([
    layers.Dense(64, activation='relu', input_shape=(num_features,)),
    layers.Dense(32, activation='relu'),
    layers.Dense(num_classes, activation='sigmoid')  # sigmoid → independent prob for each class
])
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',   # for multi-label classification
              metrics=['accuracy'])

# Train the model
history = model.fit(X, y, epochs=10, batch_size=32, verbose=2)

# Evaluate
loss, acc = model.evaluate(X, y, verbose=0)
print(f"Training Accuracy: {acc:.4f}")
```
