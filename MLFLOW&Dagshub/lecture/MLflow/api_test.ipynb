### Load Model

```python
import requests
import json

# 1. Update the port to 5001 (where your server is now running)
url = "http://127.0.0.1:5001/invocations"

# 2. Update the data format. 
# For MLflow, 'inputs' is the most common format for list-based data.
data = {
    "inputs": [[5.1, 3.5, 1.4, 0.2]] 
}

try:
    # 3. Send the POST request
    resp = requests.post(url, json=data)
    
    # 4. Check if the request was successful
    if resp.status_code == 200:
        print("Prediction:", resp.json())
    else:
        print(f"Error {resp.status_code}: {resp.text}")

except requests.exceptions.ConnectionError:
    print("Error: Could not connect to the server. Is your MLflow model serving on port 5001?")
```
